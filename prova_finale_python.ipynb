```python
#confrontando nc_na_vs_count_per_iso con tc_na_vs_count_per_iso si ha una panoramica dei valori nulli in new_cases e total_cases per iso_code.
#siccome in certi casi abbiamo molti dati nulli nei total_cases, mentre in altri ne abbiamo molti in new_cases, la migliore approssimazione sulla base dei dati presenti in questo df è:
#valutare la somma dei new_cases per iso_code e il massimo dei total_cases per iso_code e prendere il valore maggiore, che sarà il più completo.
#
```

```python
iso_new_sum = covid_df.groupby(["iso_code"])["new_cases"].sum()
iso_new_sum
```

```python
iso_tot_max = covid_df.groupby(["iso_code"])["total_cases"].max()
iso_tot_max
```

```python
iso_new_sum = iso_new_sum.reset_index()
```

```python
iso_new_sum = iso_new_sum.rename(columns = {"new_cases":"cases"})
iso_new_sum
```

```python
iso_tot_max = iso_tot_max.reset_index()
```

```python
iso_tot_max = iso_tot_max.rename(columns = {"total_cases":"cases"})
iso_tot_max
```

```python
cases_per_iso = pd.concat([iso_tot_max.reset_index(),iso_new_sum.reset_index()])
max_cases_per_iso = cases_per_iso.groupby(["iso_code"])["cases"].max().reset_index()
max_cases_per_iso
```

```python
continent_per_iso = covid_df[["iso_code","continent"]].drop_duplicates().dropna()
continent_per_iso
```

```python
# Si chiede di trovare, per ogni continente: 
# a. il numero di casi fin dall'inizio della pandemia 
#RISPOSTA al punto a:
tot_cases_per_continent = pd.merge(left = max_cases_per_iso, right = continent_per_iso).groupby("continent")["cases"].sum()
tot_cases_per_continent
```

```python
tot_cases_per_continent = tot_cases_per_continent.reset_index()
```

```python
# b. la percentuale rispetto al totale mondiale del numero di casi
#RISPOSTA al punto b:
print("Percentuali rispetto al totale mondiale del numero di casi:")
for i in range(6):
    print(f" - {tot_cases_per_continent.iloc[i,0]}: {round(tot_cases_per_continent.iloc[i,1]/tot_cases_per_continent["cases"].sum()*100,1)} %")
```

```python
covid_df[IGFS_filter*_2023_filter].groupby("location")[["hosp_patients","weekly_hosp_admissions"]].sum()
# NOTA dalla fonte dei dati (https://github.com/owid/covid-19-data/tree/master/public/data)
# hosp_patients: Number of COVID-19 patients in hospital on a given day
# weekly_hosp_admissions: Number of COVID-19 patients newly admitted to hospitals in a given week (reporting date and the preceeding 6 days)
# Quindi la somma dei pazienti ospitalizzati (intesi come nuovi pazienti) per ogni nazione considerata, per anno è descritta meglio dalla somma di weekly_hosp_admissions per nazione nel 2023.
# Considerare hosp_patients comporterebbe una sovrastima dei pazienti ospedalizzati perché un dato paziente può restare più di un giorno in ospedale ed essere quindi riconteggiato ogni giorno.
# Vediamo inoltre che la Germania non presenta dati hosp_patients.
# Per tutte queste ragioni è più sensato basare l'analisi su weekly_hosp_admissions.
```
