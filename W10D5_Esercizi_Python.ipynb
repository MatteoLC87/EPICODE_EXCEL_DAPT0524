```python
#alternativa 2 bar chart
prices_df = dimp["DealerPrice"]
prices = pd.array(prices_df)

scaling_factor = 15
threshold_75perc = 722

prices_scaled = prices/scaling_factor
filter_75perc = prices_scaled>(threshold_75perc/scaling_factor)


for price in prices_scaled[filter_75perc]:
    counter = 0
    while counter < price:
        print("*", sep="", end = "")
        counter +=1
    print("")
```

```python
#Qual è il prezzo medio per i prodotti di colore blu?
color_blue_filter = dimp["Color"] == "Blue"
prices = dimp["DealerPrice"]
blue_average_price = round(prices[color_blue_filter].mean(),1)
blue_average_price
```

```python
#• Qual è il prezzo medio per i prodotti di colore rosso o nero? 
red_filter = dimp["Color"] == "Red"
black_filter = dimp["Color"] == "Black"
prices = dimp["DealerPrice"]
red_or_black_avg_price = round(prices[red_filter | black_filter].mean(),1)
red_or_black_avg_price
```

```python
#• Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg? 
dimp[["Size","Weight"]]
size_filter = dimp["Size"] == "42"
weight_filter = dimp["Weight"] > 10
dimp["DealerPrice"][size_filter & weight_filter].max()
```

```python
#• Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento?
size_filter = dimp["Size"] == "42"
weight_filter = dimp["Weight"] > 10
color_filter = dimp["Color"] == "Silver"
dimp[["EnglishProductName","StandardCost"]][size_filter & weight_filter & color_filter]
```
